```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer

nltk.download('gutenberg')
nltk.corpus.gutenberg.fileids()
```

```python
PL = nltk.corpus.gutenberg.words('milton-paradise.txt')
PL
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer(language = 'english')
stemmed_words = []
for word in PL:
    stemmed = stemmer.stem(word)
    stemmed_words.append(stemmed)

stemmed_words    
```

```python
sw = set(nltk.corpus.stopwords.words('english'))
sw
```

```python
filtered_list = [word for word in stemmed_words if word.lower() not in sw]
new_filtered_list = [word for word in filtered_list if word.isalnum()]
Meaningless_words = ['thou','thi','thee','thus']
newest_filtered_list = [word for word in new_filtered_list if word not in Meaningless_words]
```

```python
from nltk.probability import FreqDist
fdist = FreqDist()
```

```python
for word in newest_filtered_list:
    fdist[word] += 1
fdist
```

```python
top20 = fdist.most_common(20)
top20
```

```python
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
raw_book = nltk.corpus.gutenberg.raw('milton-paradise.txt')

sentences = nltk.sent_tokenize(raw_book)
sentiments = []
for sentence in sentences:
    sentiment = analyzer.polarity_scores(sentence)
    sentiments.append((sentence, 
                       sentiment['neg']))

most_negative = sorted(sentiments, key=lambda x: x[1], reverse=True)[:10]
for sentence, score in enumerate(most_negative):
    print(f"Negative Sentiment Score: {score}")
    print()
```

```python
sentiments_pos = []
for sentence in sentences:
    sentiment = analyzer.polarity_scores(sentence)
    sentiments_pos.append((sentence, sentiment['pos']))

most_positive = sorted(sentiments_pos, key=lambda x: x[1], reverse=True)[:10]
for sentence, score in enumerate(most_positive):
    print(f"Positive Sentiment Score: {score}")
    print()
```

```python
sentiments_nue = []
for sentence in sentences:
    sentiment = analyzer.polarity_scores(sentence)
    sentiments_nue.append((sentence, sentiment['neu']))

most_nuetral = sorted(sentiments_nue, key=lambda x: x[1], reverse=True)[:10]
for sentence, score in enumerate(most_nuetral):
    print(f"Neutral Sentiment Score: {score}")
    print()
```
